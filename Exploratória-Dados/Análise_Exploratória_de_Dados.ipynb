```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# --- Configuração de Caminhos e Carregamento ---
caminho_projeto = '/content/home-credit-default-risk'
imagens_dir = 'imagens'

if not os.path.exists(imagens_dir):
    os.makedirs(imagens_dir)
```

```python
df=pd.read_csv('df_train_final.csv')
```

```python
# Selecionar algumas das features criadas para a análise de correlação
features_to_correlate = df.columns
# Ajustar para o caso do merge não encontrar a coluna exata
features_to_correlate = [col for col in features_to_correlate if col in df.columns]

# Calcular a correlação com a variável TARGET
correlation = df[features_to_correlate].corr()['TARGET'].sort_values(ascending=False)

print("\nCorrelação das features com a variável TARGET:")
print(correlation)
```

```python
strong_correlations = correlation[(correlation > 0.06) | (correlation < -0.06)]
```

```python
# --- Criar o gráfico com subplots ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Plot para EXT_SOURCE_1
sns.histplot(data=df, x='EXT_SOURCE_1', hue='TARGET',
             ax=axes[0], kde=True, bins=50, palette='viridis',
             multiple='stack', legend=False)
axes[0].set_title('Distribuição de EXT_SOURCE_1 vs. Inadimplência', fontsize=14)
axes[0].set_xlabel('EXT_SOURCE_1')
axes[0].set_ylabel('Contagem')
axes[0].set_xlim(0, 1)

# Plot para EXT_SOURCE_2
sns.histplot(data=df, x='EXT_SOURCE_2', hue='TARGET',
             ax=axes[1], kde=True, bins=50, palette='viridis',
             multiple='stack', legend=False)
axes[1].set_title('Distribuição de EXT_SOURCE_2 vs. Inadimplência', fontsize=14)
axes[1].set_xlabel('EXT_SOURCE_2')
axes[1].set_ylabel('Contagem')
axes[1].set_xlim(0, 1)

# Plot para EXT_SOURCE_3
sns.histplot(data=df, x='EXT_SOURCE_3', hue='TARGET',
             ax=axes[2], kde=True, bins=50, palette='viridis',
             multiple='stack', legend=False)
axes[2].set_title('Distribuição de EXT_SOURCE_3 vs. Inadimplência', fontsize=14)
axes[2].set_xlabel('EXT_SOURCE_3')
axes[2].set_ylabel('Contagem')
axes[2].set_xlim(0, 1)

# Adicionar legenda na figura
handles, labels = axes[0].get_legend_handles_labels()
labels = ['Pago (0)', 'Inadimplente (1)']
fig.legend(handles, labels, loc='upper right', title='Status do Empréstimo', fontsize=12)

plt.tight_layout()

# Salvar o gráfico
plt.savefig(os.path.join(imagens_dir, 'ext_source_distributions.png'))
plt.close()
```

```python
# Converter DAYS_BIRTH (negativo) para Idade (positivo)
df['AGE_IN_YEARS'] = df['DAYS_BIRTH'] / -365

# Calculate the mean of EXT_SOURCE_2 for each age and target combination
ext_source_2_mean_by_age_target = df.groupby(['AGE_IN_YEARS', 'TARGET'])['EXT_SOURCE_2'].mean().reset_index()

# --- Criar o gráfico de linha ---
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(12, 8))
sns.lineplot(
    x='AGE_IN_YEARS',
    y='EXT_SOURCE_2',
    hue='TARGET',
    data=ext_source_2_mean_by_age_target,
    palette='viridis'
)

plt.title('Relação entre Idade do Cliente e Média do Score EXT_SOURCE_2', fontsize=16)
plt.xlabel('Idade do Cliente (Anos)', fontsize=12)
plt.ylabel('Média do Score EXT_SOURCE_2', fontsize=12)
plt.legend(title='Status do Empréstimo', labels=['Pago (0)', 'Inadimplente (1)'])

# Salvar o gráfico
plt.savefig(os.path.join(imagens_dir, 'age_vs_mean_ext_source_2_line.png'))
plt.close()
```

```python
# Converter DAYS_BIRTH (negativo) para Idade (positivo)
df['AGE_IN_YEARS'] = df['DAYS_BIRTH'] / -365

# Create age bins of 5 years
bins = range(int(df['AGE_IN_YEARS'].min()), int(df['AGE_IN_YEARS'].max()) + 6, 5)
labels = [f'{i}-{i+4}' for i in bins[:-1]]
df['AGE_BIN'] = pd.cut(df['AGE_IN_YEARS'], bins=bins, labels=labels, right=False, include_lowest=True)

# Calculate the mean of EXT_SOURCE_2 for each age bin and target combination
ext_source_2_mean_by_age_bin_target = df.groupby(['AGE_BIN', 'TARGET'])['EXT_SOURCE_2'].mean().reset_index()

# Convert AGE_BIN to categorical with ordered labels for correct plotting order
ext_source_2_mean_by_age_bin_target['AGE_BIN'] = pd.Categorical(ext_source_2_mean_by_age_bin_target['AGE_BIN'], categories=labels, ordered=True)

# --- Criar o gráfico de linha ---
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(14, 8))
sns.lineplot(
    x='AGE_BIN',
    y='EXT_SOURCE_2',
    hue='TARGET',
    data=ext_source_2_mean_by_age_bin_target,
    palette='viridis'
)

plt.title('Relação entre Faixa Etária do Cliente e Média do Score EXT_SOURCE_2', fontsize=16)
plt.xlabel('Faixa Etária (Anos)', fontsize=12)
plt.ylabel('Média do Score EXT_SOURCE_2', fontsize=12)
plt.legend(title='Status do Empréstimo', labels=['Pago (0)', 'Inadimplente (1)'])
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability

# Salvar o gráfico
plt.savefig(os.path.join(imagens_dir, 'age_bin_vs_mean_ext_source_2_line.png'))
plt.close()
```

```python
# Criamos uma lista com as novas colunas de educação
education_cols = [col for col in df.columns if col.startswith('NAME_EDUCATION_TYPE')]
education_cols.append('TARGET') # Adicionamos a TARGET para agrupar

# Agrupamos por TARGET e calculamos a média para obter a proporção
plot_data = df[education_cols].groupby('TARGET').sum().T.reset_index()

# O reshaping do DataFrame para o gráfico de barras
plot_data.columns = ['Education_Type', 'Pago (0)', 'Inadimplente (1)']
plot_data = plot_data.melt(id_vars='Education_Type', var_name='Target Status', value_name='QTD')
plot_data['porcentagem'] = plot_data['QTD'] / plot_data.groupby('Target Status')['QTD'].transform('sum')
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(14, 8))
ax = sns.barplot(
    x='Education_Type',
    y='porcentagem',
    hue='Target Status',
    data=plot_data,
    palette='viridis'
)

plt.title('Proporção de Clientes por Nível de Educação vs. Inadimplência', fontsize=16)
plt.xlabel('Nível de Educação', fontsize=12)
plt.ylabel('Proporção de Clientes', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Status do Empréstimo')
plt.tight_layout()

# Salvar o gráfico
plt.savefig(os.path.join(imagens_dir, 'education_vs_target_barplot_post_ohe.png'))
plt.close()
```

```python
# Criamos uma lista com as novas colunas de educação
family_cols = [col for col in df.columns if col.startswith('NAME_FAMILY_STATUS')]
family_cols.append('TARGET') # Adicionamos a TARGET para agrupar

# Agrupamos por TARGET e calculamos a média para obter a proporção
plot_data = df[family_cols].groupby('TARGET').sum().T.reset_index()

# O reshaping do DataFrame para o gráfico de barras
plot_data.columns = ['FAMILY_STATUS', 'Pago (0)', 'Inadimplente (1)']
plot_data = plot_data.melt(id_vars='FAMILY_STATUS', var_name='Target Status', value_name='QTD')
plot_data['porcentagem'] = plot_data['QTD'] / plot_data.groupby('Target Status')['QTD'].transform('sum')
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(14, 8))
ax = sns.barplot(
    x='FAMILY_STATUS',
    y='porcentagem',
    hue='Target Status',
    data=plot_data,
    palette='viridis'
)

plt.title('Distribuição de Renda por Estado Civil vs. Inadimplência', fontsize=16)
plt.xlabel('Nível de Educação', fontsize=12)
plt.ylabel('Proporção de Clientes', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Status do Empréstimo')
plt.tight_layout()

# Salvar o gráfico
plt.savefig(os.path.join(imagens_dir, 'family_status_vs_target_barplot_post_ohe.png'))
plt.close()
```
